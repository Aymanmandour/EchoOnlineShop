```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('datasetCSV5.csv')

# Check the first few rows
print(df.head())

# Get basic statistics
print(df.describe())
```

```python
# Convert date and resolved columns to datetime
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df['resolved'] = pd.to_datetime(df['resolved'], format='%Y-%m-%d')

# Create a column for the number of days it took to resolve the issue
df['time_to_resolve'] = (df['resolved'] - df['date']).dt.days
```

```python
# Extract day, month, and year as features
df['day'] = df['date'].dt.day
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year

# Calculate time since last failure for each asset
df = df.sort_values(by=['asset', 'date'])
df['time_since_last_failure'] = df.groupby('asset')['date'].diff().dt.days
```

```python
# Define features (X) and target (y)
X = df[['asset', 'location', 'day', 'month', 'year', 'time_since_last_failure']]
y = df['failure']

# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the Random Forest model
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)
```

```python
import numpy as np
from datetime import datetime, timedelta
import itertools

# Step 1: Generate a list of dates for the next 30 days
today = datetime.today()
next_30_days = pd.date_range(today, today + timedelta(days=30))

# Step 2: Get a list of unique assets and locations from the dataset
unique_assets = df['asset'].unique()
unique_locations = df['location'].unique()

# Step 3: Create a Cartesian product of dates and assets
future_combinations = list(itertools.product(next_30_days, unique_assets, unique_locations))

# Step 4: Create a DataFrame from these combinations
future_df = pd.DataFrame(future_combinations, columns=['date', 'asset', 'location'])

# Step 5: Extract day, month, and year as features
future_df['day'] = future_df['date'].dt.day
future_df['month'] = future_df['date'].dt.month
future_df['year'] = future_df['date'].dt.year

# Step 6: Fill 'time_since_last_failure' with an appropriate value, e.g., 0 or estimated
future_df['time_since_last_failure'] = 0  # Fill this column appropriately

# Now the future_df has the same length for all columns and is ready for predictions.
```

```python
# Use the trained model to predict failures in the next 30 days
future_df['predicted_failure'] = rf_model.predict(future_df[['asset', 'location', 'day', 'month', 'year', 'time_since_last_failure']])

# Show the predicted failures
print(future_df[['date', 'asset', 'location', 'predicted_failure']])
```

```python
# Step 1: Apply the trained model to the entire dataset (past and future)
# First, predict for past failures in the original dataset (df)
df['predicted_failure'] = rf_model.predict(df[['asset', 'location', 'day', 'month', 'year', 'time_since_last_failure']])

# Step 2: Merge predicted failures with actual failure types from the original dataset
# This will include both actual past failures and predicted future failures
df_with_failure_types = pd.merge(df, df[['asset', 'failure']], on='asset', how='left')

# Step 3: Apply the same prediction process to future data (next 30 days)
# Use the future_df to predict failures for the future dates
future_df['predicted_failure'] = rf_model.predict(future_df[['asset', 'location', 'day', 'month', 'year', 'time_since_last_failure']])

# Step 4: Merge future predictions with the original dataset to bring in failure types
future_df_with_types = pd.merge(future_df, df[['asset', 'failure']], on='asset', how='left')

# Step 5: Combine the past data and future predictions into one DataFrame
combined_df = pd.concat([df_with_failure_types, future_df_with_types], ignore_index=True)

# Step 6: Filter to only rows where a failure is predicted (predicted_failure == 1)
predicted_failures_all = combined_df[combined_df['predicted_failure'] == 1]

# Step 7: Select the relevant columns and drop duplicates
top_30_failures = predicted_failures_all[['asset', 'failure']].drop_duplicates().head(30)

# Rename the columns for the report
top_30_failures.columns = ['Asset', 'Predicted Failure Type']

# Step 8: Display the top 30 assets and their predicted failure types as a table
print("Top 30 Assets and Their Predicted/Actual Failure Types Across All Data:")
print(top_30_failures.to_string(index=False))


```
